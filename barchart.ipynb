```python
# import required packages
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Here is the dataset
df = pd.read_csv(r"C:\Users\GH414LF\Downloads\Create 2D charts\barcharts-us-music-sales-by-genre.csv" ,index_col='Year')
df.head()
```

```python
# a bar chart of Jazz vs. Rock to compare their shares in the US market
fig, ax = plt.subplots(figsize=(12,6))

rock = ax.bar(df.index, 'Rock', data=df, label='Rock %')
jazz = ax.bar(df.index, 'Jazz', data=df, label='Jazz %')

ax.legend()

```

```python
# add a title to the chart
ax.set_title('Rock vs. Jazz', size=22)

ax.set_xticks(df.index)

spines=['bottom','top','left','right']

for s in spines:
    ax.spines[s].set_visible(False)
    
ax.legend(frameon=False)

ax.tick_params(left='off', bottom='off')
    
fig     
```

```python
for p in rock.patches:
    t = ax.annotate(str(p.get_height())+'%', (p.get_x(), p.get_height()+2))
    t.set(color='black', size=8)
    
    t.set_bbox(dict(facecolor='white', boxstyle='circle',lw=1, edgecolor='red'))

fig
```

```python
# Instantiate figure and axes
fig, ax = plt.subplots(figsize=(15,5))


# The top most bar could be, say, Jazz
ax.bar(df.index, df['Jazz'], bottom=df['Rock']+df['Country'], label='Jazz')
# and it sits on Rock and Country Y values
# You define it by setting bottom value

# One below, could be Rock, and it sits on Country only
ax.bar(df.index, df['Rock'], bottom=df['Country'], label='Rock')

# and the last bars belong to Country and it sits on nothing
ax.bar(df.index, df['Country'], label='Country')

# Let's show all the index values as they are seen in the dataframe
ax.set_xticks(df.index)


ax.legend()
```
